```python
import re
from collections import Counter
import math
```

```python
file_name = "CryptoText.txt"
file_name2 = "CleanedWithSpaces.txt"
file_name3 = "CleanedWithoutSpaces.txt"
```

```python
try:
    with open(file_name, 'r', encoding='utf-8') as file:
        start_text = file.read().lower().replace('\n', '')
        # start_text = start_text.replace('ё', 'е').replace('ъ', 'ь')
        cleaned_text = re.sub(r'[^а-яё ]', '', start_text)
        with_spaces = " ".join(cleaned_text.split()) #Текст з пробілами
        without_spaces = "".join(cleaned_text.split()) #Текст без пробілів

    with open(file_name2, 'w', encoding='utf-8') as file2:
        file2.write(with_spaces)
        file2.close()

    with open(file_name3, 'w', encoding='utf-8') as file3:
        file3.write(without_spaces)
        file3.close()

except Exception as e:
    print(e)
```

```python
def entropy_count(input_text, h="mono"):
    if h == "mono":
        text_ngrams = input_text
        h = 1
    elif h == "bigrams1":
        h = 2
        text_ngrams = [input_text[i:i+2] for i in range(len(input_text) - 1)]
    elif h == "bigrams2":
        text_ngrams = [input_text[i:i+2] for i in range(0, len(input_text) - 1, 2)]
        h = 2
    else:
        return

    counts = dict(Counter(text_ngrams))
    total_length = len(text_ngrams)


    entropy = 0
    chance_dict = {}
    for char, count in counts.items():
        p = count / total_length
        contribution = p * math.log2(p)
        entropy -= contribution
        chance_dict[char] = p

    entropy = entropy/h

    alphabet = set("".join(counts.keys()))
    print(alphabet)
    print(len(alphabet))
    total_entropy = math.log2(len(alphabet))
    print(total_entropy)
    redundancy = 1 - (entropy/total_entropy)
    print(redundancy)

    sorted_ngrams = sorted(counts.items())
    sorted_chances = sorted(chance_dict.items())
    sorted_n, sorted_p = {}, {}


    for char, count in sorted_ngrams:
        sorted_n[char] = count

    for char, count in sorted_chances:
        sorted_p[char] = count

    print(sorted_n, "\n")
    print(sorted_p, "\n")
    print(f"Ентропія: {entropy}")
    print(f"Надлишковість: {redundancy}")
```

```python
entropy_count(with_spaces) #З проб.
```

```python
entropy_count(without_spaces) #Без проб.
```
