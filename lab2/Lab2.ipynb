```python
def encrypt(key, key_length):
    key_stream = (key * (text_length // key_length)) + key[:text_length % key_length]
    ciphertext = []
    # print(key_stream, len(key_stream), text_length)

    # for i, text_char in enumerate(task1):
    for text_char, key_char in zip(task1, key_stream):
        x_i = letter_num[text_char]

        #key_char = key[i % key_length] #ki = ki mod r
        k_i = letter_num[key_char]

        encrypted_char = (x_i + k_i) % alphabet_length
        ciphertext.append(num_letter[encrypted_char])

    encrypted = "".join(ciphertext)

    filename = f"./Encrypted/c_text_r{key_length}.txt"

    with open(filename, 'w', encoding='utf-8') as file3:
        file3.write(encrypted)

    return encrypted
```

```python
def calculate_ic(text):
    n = len(text)
    counts = Counter(text)
    part1 = sum(n_t * (n_t - 1) for n_t in counts.values())
    part2 = n*(n-1)

    return part1/part2 #ic
```

```python
n = len(task2)

def first_method(text):
    average_ic = {}

    for r in range(2, 31):
        blocks = [""] * r

        for i in range(n):
            char = text[i]
            block_number = i % r
            blocks[block_number] += char

        blocks_ic = []

        for b in blocks:
            blocks_ic.append(calculate_ic(b))

        avg_ic = sum(blocks_ic) / len(blocks_ic)
        average_ic[r] = avg_ic

    best_r = -1
    smallest = float('inf')

    for r, ic in average_ic.items():
        diff = abs(ic - theory_ic)
        print(f"R{r}, ic {ic}: {diff}")

        if diff < smallest:
            smallest = diff
            best_r = r

    print(f"r: {best_r}")
    print(f"Найменше відхилення: {smallest}")

    return best_r, smallest
```

```python
def find_key(text, r):
    blocks = [""] * r

    for i, char in enumerate(text):
        blocks[i % r] += char

    found_key = []

    for i, block in enumerate(blocks):
        block_count = Counter(block)

        y_common = block_count.most_common(1)[0][0]

        y_num_com = letter_num[y_common]
        x_num_com = letter_num["а"]

        key_i_num = (y_num_com - x_num_com) % alphabet_length
        key_i_char = num_letter[key_i_num]

        found_key.append(key_i_char)

    found_key = "".join(found_key)
    print(f"R:{r} Key:{found_key}")
    return found_key
```

```python
def vigenere_decrypt(text, key):
    decrypted = []
    key_length = len(key)

    for i, char in enumerate(text):
        y_i = letter_num[char]

        key_char = key[i % key_length]
        key_i_num = letter_num[key_char]
        decrypted_num = (y_i - key_i_num) % alphabet_length
        decrypted.append(alphabet[decrypted_num])

    return "".join(decrypted)
```
